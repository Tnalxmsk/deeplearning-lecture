```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
dsa_data = pd.read_csv('/Users/minsu/Desktop/DSA_features.csv')
dsa_data
```

## 데이터 전처리 및 분리

### 정상 비정상 분류

```python
lying_data = dsa_data[dsa_data['activity'].str.contains('lying')].drop(columns=['activity', 'people'])
jumping_data = dsa_data[dsa_data['activity'].str.contains('jumping')].drop(columns=['activity', 'people'])
```

```python
lying_data
```

```python
jumping_data
```

### lying + jumping Test Data 생성

```python
test_data = pd.concat([lying_data, jumping_data])
```

```python
test_data
```

### 표준화

```python
scaler = StandardScaler()
lying_data_scaled = scaler.fit_transform(lying_data)
test_data_scaled = scaler.transform(test_data)
```

## AutoEncoder Model create

### lying data만 갖고 훈련 AE 구축

```python
X_train, X_val = train_test_split(lying_data_scaled, test_size=0.2, random_state=42)
```

```python
# Build the Autoencoder
input_dim = X_train.shape[1]  # Number of features
encoding_dim = 64  # Dimension of the encoded representation
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Autoencoder model
autoencoder = Sequential([
    Dense(encoding_dim, activation='relu', input_shape=(input_dim,)),
    Dense(input_dim, activation='sigmoid')
])
```

```python
autoencoder.compile(optimizer='adam', loss='mse')

autoencoder.summary()
```

```python
history = autoencoder.fit(X_train, X_train, epochs=50, batch_size=32, validation_data=(X_val, X_val), verbose=0)
```

## 재구성 오차 계산

```python
# Reconstruction error on the test set (lying + jumping)
test_reconstructions = autoencoder.predict(test_data_scaled)
reconstruction_errors = np.mean(np.square(test_data_scaled - test_reconstructions), axis=1)
```

## 임계값 설정 및 분류

```python
# Set a threshold based on training data (using reconstruction error)
threshold = np.percentile(reconstruction_errors, 95)
```

```python
# Classify test data based on reconstruction error
test_labels = np.array(['lying'] * len(lying_data) + ['jumping'] * len(jumping_data))
predicted_labels = np.where(reconstruction_errors > threshold, 'jumping', 'lying')
```

## 정확도 계산

```python
# Calculate accuracy
accuracy = np.mean(test_labels == predicted_labels)

print(f'Accuracy: {accuracy:.2f}')
```
